```python
#Imports
import pandas as pd
import numpy as np

#Load data
basics_url = "https://datasets.imdbws.com/title.basics.tsv.gz"
ratings_url = "https://datasets.imdbws.com/title.ratings.tsv.gz"
akas_url = "https://datasets.imdbws.com/title.akas.tsv.gz"

basics = pd.read_csv(basics_url, sep='\t', low_memory=False)
ratings = pd.read_csv(ratings_url, sep='\t', low_memory=False)
akas = pd.read_csv(akas_url, sep='\t', low_memory=False)
```

Basics Data Clean

```python
basics.head()
```

```python
basics.replace({'\\N':np.nan},inplace=True)
basics
```

```python
basics['primaryTitle'].isna().sum()
```

```python
basics = basics.dropna(subset= ["primaryTitle"])
```

```python
basics['runtimeMinutes'].isna().sum()
```

```python
basics['genres'].isna().sum()
```

```python
basics = basics.dropna(subset=["runtimeMinutes", "genres"])
```

```python
movies = basics["titleType"] == 'movie'
```

```python
movies_filter = basics[movies]
movies_filter
```

```python
movies_filter['startYear'] = movies_filter['startYear'].astype(float)
```

```python
movies_filter.dtypes
```

```python
year = movies_filter['startYear'] >= 2000.0 
years = movies_filter['startYear'] <= 2021.0
```

```python
filtered_combined = year & years
```

```python
movie_years = movies_filter[filtered_combined]
movie_years
```

```python
is_documentary = movie_years['genres'].str.contains('documentary',case=False)
years_filtered = movie_years[~is_documentary]
```

AKAs Data Clean

```python
akas.replace({'\\N':np.nan},inplace=True)
```

```python
akas.head()
```

```python
region = akas['region'] == 'US'
```

```python
filt_akas = akas[region]
filt_akas
```

```python
filt_akas_cleaned = filt_akas
```

```python
import os 
os.makedirs('Data/',exist_ok= True)
os.listdir('Data/')
```

```python
filt_akas_cleaned.to_csv('Data/filt_akas_cleaned.csv.gz',compression='gzip', index= False)
```

```python
filt_akas_cleaned= pd.read_csv('Data/filt_akas_cleaned.csv.gz',low_memory = False)
filt_akas_cleaned.head()
```

Ratings Data Clean

```python
ratings.replace({'\\N':np.nan},inplace=True)
```

```python
ratings.head()
```

```python
keepers = ratings['tconst'].isin(filt_akas_cleaned['titleId'])
```

```python
ratings = ratings[keepers]
ratings
```

```python
filt_ratings = ratings
```

```python
filt_ratings.to_csv("Data/filt_ratings.csv.gz",compression='gzip',index=False)
```
